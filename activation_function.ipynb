```python
import numpy as np
import matplotlib.pyplot as plt

def heaviside(x):
    return np.where(x >= 0, 1, 0)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

x = np.linspace(-5, 5, 400)

plt.figure(figsize=(10, 6))
plt.plot(x, heaviside(x), label="Heaviside")
plt.plot(x, sigmoid(x), label="Sigmoïde")
plt.plot(x, tanh(x), label="Tanh")
plt.plot(x, relu(x), label="ReLU")
plt.title("Fonctions d'activation sur le même subplot")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-3, 3)
plt.ylim(-1.5, 1.5)
plt.legend()
plt.grid()
plt.show()
```
